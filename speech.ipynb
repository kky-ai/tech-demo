##### TTS: text to wav file
---

- ```example_data/test.wav``` is created using the TTS engine
- password will be provided

````python
import requests

def text2wav(text, voice, url, username, password):

    params = {
        'engine': voice,
        'text': text,
        'format': 'wav',
    }

    try:
        response = requests.get(
            url, 
            auth=requests.auth.HTTPDigestAuth(username, password), 
            params=params
        )
        
        if response.status_code == 200:
            with open('/content/test.wav', 'wb') as f:
                f.write(response.content)
                
            print("GET request successful. Response saved.")
        else:
            print("GET request failed with status code:", response.status_code)
            
    except requests.exceptions.RequestException as e:
        print("Error making GET request:", e)
````

````python
TEXT = '''
    This is an example audio file. It was generated using the text-to-speech technology 
    from the Department of Cybernetics at the University of West Bohemia in Pilsen, Czech Republic. 
    May the force be with you.
    '''

## Czech
#VOICE = 'Oldrich30'
#VOICE = 'Ilona30'

## English    
VOICE = 'Tim30'
#VOICE = 'Emma30'

## French
#VOICE = 'Sophie30'
#VOICE = 'Frederik30'

## Russian
#VOICE = 'Alexej30'
#VOICE = 'Julija30'

## German
#VOICE = 'Stephanie30'

text2wav(
    text=TEXT,
    voice=VOICE,
    url='https://ryzen2.megaword.cz:9999/tts-c4dhi/v4/synth',
    username='c4dhi',
    password=''
)
````

````python
import requests

LANGUAGE = 'en'
API_URL = f'https://uwebasr.zcu.cz/api/v2/lindat/malach/{LANGUAGE}'
FILE_PATH = '/content/test.wav'

with open(FILE_PATH, 'rb') as fr:
    r = requests.post(API_URL, data=fr, params={"format": "json"})
    r.raise_for_status()
    ret = r.json()
    
print(' '.join([i['word'] for i in ret]))
````
